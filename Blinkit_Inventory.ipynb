```python
import numpy as np # as scientific and mathematical calculation
import pandas as pd # as Data Manipulation.
import seaborn as sns # as visualization.
import matplotlib.pyplot as plt # as visualization
from mpl_toolkits.mplot3d import Axes3D # 3D plotting
```

```python
orders = pd.read_csv(r"F:\pythonProject1\Blinkit Inventory and Sply chain\Orders.csv")
orders.head()
```

```python
orders.info()
```

```python
orders["OrderDateTime"] = pd.to_datetime(orders["OrderDateTime"])
orders.head()
```

```python
orders["ExpectedDeliveryTime"] = pd.to_datetime(orders["ExpectedDeliveryTime"])
orders["ActualDeliveryTime"] = pd.to_datetime(orders["ActualDeliveryTime"])
orders.head()
```

```python
orders["Actual_Delivery_in_mins"] = (orders["ActualDeliveryTime"] - orders["OrderDateTime"]).dt.total_seconds()/60
orders.head()
```

```python
orders["Expected_Delivery_in_mins"] = (orders["ExpectedDeliveryTime"] - orders["OrderDateTime"]).dt.total_seconds()/60
orders.head()
```

```python
# to see Distinct values in a column
orders["StoreID"].unique()
```

```python
orders["Diff_bw_Expected_Actual"] = orders["Expected_Delivery_in_mins"] - orders["Actual_Delivery_in_mins"]
orders.head()
```

```python
#loc allows us to access a group of rows and columns by label(s) or a boolean array.
# iloc is integer index based, so you have to specify rows and columns by their integer position values (0-based integer position). 
orders.loc[(orders["Diff_bw_Expected_Actual"] == 0), "orders_status"] = "On Time"
orders.loc[(orders["Diff_bw_Expected_Actual"] < 0), "orders_status"] = "Early"
orders.loc[(orders["Diff_bw_Expected_Actual"] > 0), "orders_status"] = "Delayed"
orders.head()
```

```python
plt.figure(figsize = (16,4))
sns.countplot(x ="StoreID" , data= orders , hue = "orders_status")
plt.xticks(rotation = 90)
plt.show()
```

```python
products = pd.read_csv(r"F:\pythonProject1\Blinkit Inventory and Sply chain\Products.csv") 
products.head()
```

```python
df = pd.merge(left = orders, right = products, how = "left", on = "ProductID")
df.head()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Aggregate data: total amount by Category and OrderStatus
agg = df.groupby(['Category', 'OrderStatus'])['TotalAmount'].sum().reset_index()

# Create 3D figure
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

# Map categories and statuses to numeric positions
categories = agg['Category'].unique()
statuses = agg['OrderStatus'].unique()
cat_idx = {cat: i for i, cat in enumerate(categories)}
status_idx = {status: i for i, status in enumerate(statuses)}

# Plot bars
for i, row in agg.iterrows():
    x = cat_idx[row['Category']]
    y = status_idx[row['OrderStatus']]
    z = 0
    dx = 0.5
    dy = 0.5
    dz = row['TotalAmount']
    ax.bar3d(x, y, z, dx, dy, dz, shade=True)

# Label axes
ax.set_xticks(list(cat_idx.values()))
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.set_yticks(list(status_idx.values()))
ax.set_yticklabels(statuses)
ax.set_zlabel('TotalAmount')
ax.set_xlabel('Category')
ax.set_ylabel('Order Status')
ax.set_title('3D Total Amount by Category and Order Status')

plt.show()
```
